```python
import pandas as pd
from functools import reduce

# Load all datasets
df_list = [
    pd.read_csv("normalized_pose3d_dataset.csv"),
    pd.read_csv("pose3d_distance_features.csv"),
    pd.read_csv("pose3d_symmetry_features.csv"),
    pd.read_csv("pose_joint_angles_dataset.csv"),
    pd.read_csv("pose3d_velocity_features.csv"),
    pd.read_csv("pose3d_acceleration_features.csv")
]

# Merge on 'video_name' and 'frame_id'
df_merged = reduce(lambda left, right: pd.merge(left, right, on=['video_name', 'frame_id'], how='outer'), df_list)

# Fill any remaining NaNs
df_merged.fillna(0, inplace=True)
```

```python
# Reviewing the merged dataset
df_merged.head()
```

```python
# Saving the data
df_merged.to_csv("final_pose_features_dataset.csv", index=False)
```
